```python
# Import the packages we will use in this project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Read in the data

df = pd.read_csv('/Users/jasminechung/Downloads/movies.csv')
```

```python
# Loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100,3)))
```

```python
# Replace NaNs with 0s in budget and gross columns

df['budget'] = df['budget'].fillna(0)
df['gross'] = df['gross'].fillna(0)
df
```

```python
# Data types for columns

print(df.dtypes)
```

```python
# Change budget and gross columns data type into int

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# Order the data by gross column

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Extract year released from release column

dayreleased = df['released'].str.split('(',expand=True)
yearreleased = dayreleased[0].str.split(',',expand=True)
df['yearcorrect'] = yearreleased[1]
df
```

```python
# Drop duplicates

df.drop_duplicates()
```

```python
# Look at the correlation between budget and gross using scatter plot

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for film')
plt.ylabel('Gross earnings')
plt.show()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
# Look at correlation using pearson method

df.corr()
```

```python
# Look at correlation using heatmap

correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for numeric features')
plt.xlabel('Movie factors')
plt.ylabel('Movie factors')
plt.show()
```

```python
# Look at company's correlation to gross using a numerized column

df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized
```

```python
df
```

```python
# Look at correlation using heatmap again with numerized df

correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for numeric features')
plt.xlabel('Movie factors')
plt.ylabel('Movie factors')
plt.show()
```

```python
# Unstack correlation and see which factors highly correlate

correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
sorted_pairs = corr_pairs.sort_values()
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
